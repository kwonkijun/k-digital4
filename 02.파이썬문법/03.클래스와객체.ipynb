```python
class Monster:
    def __init__(self, name, attack):
        self.name = name
        self.attack = attack
    def info(self):
        print(f'이름: {self.name} 공격력: {self.attack}')

lion = Monster("사자", 100)
baron = Monster("바론", 240)
lion.info()
baron.info()
```

```python
# 캐릭터 클래스를 만들고, 속성 메서드 사용해보기
class Character:
    def __init__(self, name, level, hp):
        self.name = name
        self.level = level
        self.hp = hp
    def info(self):
        print('------캐릭터 정보------')
        print(f'이름 : {self.name}')
        print(f'레벨 : {self.level}')
        print(f'체력 : {self.hp}')

ezreal = Character('이즈리얼', 1, 600)
ezreal.level += 1
ezreal.hp += 150
ezreal.say = "내 솜씨를 제대로 보여줄 시간이군"
print(ezreal.say)
```
